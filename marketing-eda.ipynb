```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
bank_marketing = fetch_ucirepo(id=222) 
  
# data (as pandas dataframes) 
X = bank_marketing.data.features 
y = bank_marketing.data.targets 
  
# metadata 
print(bank_marketing.metadata) 
  
# variable information 
print(bank_marketing.variables) 
```

## Exploratory Data Analysis

### Univariate Analysis

```python
X['Outcome']=y
```

```python
numeric=['age', 'balance', 'duration', 'campaign','pdays', 'previous']
for i in numeric:
    plt.figure(figsize=(8,5))
    plt.subplot(1,2,1)
    X[ X['Outcome'] == 'yes'][i].plot(kind='box',colormap='viridis')
    plt.xlabel(f'{i} for yes')
    plt.ylabel('Distribution')
    
    plt.subplot(1,2,2)
    X[ X['Outcome'] == 'no'][i].plot(kind='box')
    plt.xlabel(f'{i} for no')
    plt.ylabel('Distribution')
    plt.show()
```

```python
categorical=['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact','poutcome']

for i in categorical:
    plt.figure(figsize=(8,5))
    plt.subplot(1,2,1)
    X[ X['Outcome'] == 'yes'][i].value_counts().plot(kind='bar',colormap='viridis')
    plt.xlabel(f'{i}-type for yes')
    plt.ylabel('Counts')
    
    plt.subplot(1,2,2)
    X[ X['Outcome'] == 'no'][i].value_counts().plot(kind='bar')
    plt.xlabel(f'{i}-type for no')
    plt.ylabel('Counts')
    plt.show()
```

```python
y.value_counts().plot(kind='pie',autopct='%1.1f')
plt.show()
```

### Bivariate Analysis

```python
mapping_dict = {'no':0, 'yes': 1}
X.loc[:,'Outcome']=X['Outcome'].map(mapping_dict)
```

```python
yes_depo=X[ X['Outcome'] == 1]
no_depo=X[ X['Outcome'] == 0]

pd.crosstab(yes_depo['marital'],values=yes_depo['job'],columns=yes_depo['job'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=Yes')

pd.crosstab(no_depo['marital'],values=no_depo['job'],columns=no_depo['job'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=No')
plt.show()
```

```python

pd.crosstab(yes_depo['poutcome'],values=yes_depo['loan'],columns=yes_depo['loan'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=Yes')

pd.crosstab(no_depo['poutcome'],values=no_depo['loan'],columns=no_depo['loan'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=No')
plt.show()
```

```python

pd.crosstab(yes_depo['contact'],values=yes_depo['job'],columns=yes_depo['job'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=Yes')

pd.crosstab(no_depo['contact'],values=no_depo['job'],columns=no_depo['job'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=No')
plt.show()
```

```python
pd.crosstab(yes_depo['housing'],values=yes_depo['default'],columns=yes_depo['default'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=Yes')

pd.crosstab(no_depo['housing'],values=no_depo['default'],columns=no_depo['default'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=No')
plt.show()
```

```python
pd.crosstab(yes_depo['loan'],values=yes_depo['default'],columns=yes_depo['default'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=Yes')

pd.crosstab(no_depo['loan'],values=no_depo['default'],columns=no_depo['default'],aggfunc='count',normalize=True).plot(kind='bar')
plt.title('For deposited=No')
plt.show()
```
